Bu veri seti, tümörlerini aldırmak için ameliyat olan bir grup meme kanseri hastasından oluşmaktadır. Veri kümesi aşağıdaki değişkenlerden oluşur:

#### Veri İçeriği
Patient_ID: bir hastanın benzersiz tanımlayıcı kimliği

Age: tanı yaşı (Yıl)

Geder: Male(Erkek)  Female(Kadın)

Protein1, Protein2, Protein3, Protein4: ekspresyon seviyeleri (tanımsız birimler)

Tumour_Stage(Tümör_Evresi): I, II, III

Histology: İnfiltratif Duktal Karsinom, İnfiltratif Lobüler Karsinom, Müsinöz Karsinom

ER durumu: Olumlu/Negatif

PR durumu: Olumlu/Negatif

HER2 durumu: Pozitif/Negatif

Surgery_type(Cerrahi_tipi): Lumpektomi, Basit Mastektomi, Modifiye Radikal Mastektomi, Diğer

Date_of_Surgery: Ameliyatın gerçekleştirildiği tarih (GG-MON-YY olarak)

Date_of_Last_Visit: Son ziyaret tarihi (GG-MON-YY olarak) [hastanın ameliyattan sonra tekrar ziyaret etmemesi durumunda boş olabilir]

Patient_Status: Hayatta/Ölü [ameliyattan sonra hastanın tekrar ziyaret etmemesi ve hastanın hayatta olup olmadığına dair bilgi olmaması durumunda boş olabilir].
